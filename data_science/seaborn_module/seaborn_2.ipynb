### lecture 2 - categorical plots

```python
import seaborn as sns
import numpy as np
```

```python
tips = sns.load_dataset('tips')
```

#### barplot

```python
# here we will get the mean of the total bill column per cateogry
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
# np.std will tell me what is the standard deviation of the total bill column per category
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

#### countplot

```python
sns.countplot(x='sex',data=tips)
```

#### boxplots and violin plots

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.violinplot(x='sex',y='total_bill',data=tips)
```

#### strip plot

```python
# jitter is going to add a little bit of random noise in order to separate overlapping points
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex')
```

#### swarm plot

combination of a strip plot and a violin plot

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```
