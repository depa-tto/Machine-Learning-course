### Pandas Data Visualization Exercise

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df3 = pd.read_csv('./dataset/df3')
```

```python
df3.info()
```

```python
df3.head()
```

recreate this scatter plot of b vs a. Note the color and size of the points. Also note the figure size

```python
# 's' stands for shape while 'c' for color
df3.plot.scatter(x='a',y='b',s=50,c='red',figsize=(12,3))
```

create a histogram of the 'a' column

```python
df3['a'].plot.hist(bins=20)
```

```python
plt.style.use('ggplot')
```

```python
df3['a'].plot.hist(bins=20,alpha=0.5)
```

create a boxplot comparing the a and b columns

```python
df3[['a','b']].boxplot()
```

create a kde plot of the 'd' column 

```python
df3['d'].plot.kde()
```

figure out how to increase the linewidth and make the linestyle dashed

```python
df3['d'].plot.kde(lw=5,ls='--')
```

create an area plot of all the columns for just the rows up to 30. (hint: use .ix)

```python
df3.loc[0:30].plot.area(alpha=0.5,cmap='summer')
```

```python
f = plt.figure()
df3.loc[0:30].plot.area(alpha=0.5,cmap='summer')
plt.legend(loc='center left',bbox_to_anchor=(1.0,0.5))
plt.show()
```
