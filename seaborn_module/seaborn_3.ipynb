### lecture 3 - matrix plots

```python
import numpy as np
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
flights = sns.load_dataset('flights')
```

```python
tips.head()
```

```python
flights.head()
```

```python
numerical_tips = tips.select_dtypes(include=[np.number])
```

```python
tc = numerical_tips.corr()
```

#### heatmap

```python
# annot will annotate each numerical value for each color
sns.heatmap(tc,cmap='summer',annot=True)
```

```python
fp = flights.pivot_table(index='month',columns='year',values='passengers',observed=False)
```

```python
fp
```

```python
sns.heatmap(fp,cmap='summer',linecolor='white',linewidths=1)
```

```python
sns.heatmap(fp,cmap='magma')
```

```python
sns.heatmap(fp,cmap='coolwarm')
```

#### cluster map

```python
sns.clustermap(fp,cmap='coolwarm')
```

```python
# standard_scale=1 will normalize the results from 0 to 1
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
```
