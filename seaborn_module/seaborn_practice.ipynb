```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

#### exercises

```python
sns.jointplot(x='fare',y='age',data=titanic)
```

```python
sns.histplot(x='fare',data=titanic,kde=True,color='red',bins=30)
```

```python
sns.boxplot(x='class',y='age',data=titanic)
```

```python
sns.countplot(x='sex',data=titanic)
```

```python
numerical_titanic = titanic.select_dtypes(include=[np.number])
```

```python
tc = numerical_titanic.corr()
```

```python
sns.heatmap(tc,cmap='coolwarm')
plt.title('tatanic variables correlation')
```

```python
g = sns.FacetGrid(data=titanic,col='sex')
g.map(plt.hist,'age')
```

```python
g = sns.FacetGrid(data=titanic,col='sex')
g.map(sns.histplot,'age')
```
