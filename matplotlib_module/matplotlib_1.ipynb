### lecture 1

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
```

```python
y = x**2
```

```python
x
```

```python
y
```

#### functional method of plotting

```python
plt.plot(x,y,'red')
```

```python
plt.plot(x,y)
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'red')

plt.subplot(1,2,2)
plt.plot(y,x,'blue')
```

#### object oriented method of plotting

```python
fig = plt.figure() # figure object as a blank canvas

axes = fig.add_axes([0.1,0.1,0.8,0.8]) # axes

axes.plot(x,y)
axes.set_xlabel('x label')
axes.set_ylabel('y label')
axes.set_title('title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) 
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) # left, bottom, width and height with respect to the empty canva
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) 
axes2 = fig.add_axes([0.4,0.15,0.4,0.3])
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) 
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('larger plot')

axes2.plot(y,x)
axes2.set_title('smaller plot')
```

### lecture 2

```python
fig,axes = plt.subplots(nrows=3, ncols=2)

# axes.plot(x,y)
plt.tight_layout() # fix the overlap of subplots
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

for current_ax in axes: # axes object is iterable 
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_title('first plot')

axes[1].plot(y,x)
axes[1].set_title('second plot')
```

#### figure size and DPI

```python
# tuple is the width and height in inches and dpi are the pixels for inch
fig = plt.figure(figsize=(3,2), dpi=100) 

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig = plt.figure(figsize=(8,2), dpi=100) 

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2), dpi=100) 

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout() # not overlapping subplots
```

#### legend

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

# label is the string that the legend function is able to reference 
ax.plot(x,x**2, label='x squared') 
ax.plot(x,x**3, label = 'x cubed')

# check if there is a label, that will be returned in the plot
ax.legend(loc='best') # best location for our legend
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

# label is the string that the legend function is able to reference 
ax.plot(x,x**2, label='x squared') 
ax.plot(x,x**3, label = 'x cubed')

# check if there is a label, that will be returned in the plot
ax.legend(loc=(0.1,0.1)) 
```
