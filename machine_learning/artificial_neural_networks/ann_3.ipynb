#### train and test

```python
X = df.drop('benign_0__mal_1',axis=1).values
```

```python
y = df['benign_0__mal_1'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

#### scaling data

```python
scaler = MinMaxScaler()
```

```python
X_train =  scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

#### model

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))

# sigmoid since is a binary classification
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test))
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

#### early stopping

```python
model = Sequential()

model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))

model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
early_stop = EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=25)
```

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test),
          callbacks=[early_stop])
```

```python
model_loss = pd.DataFrame(model.history.history)
```

```python
model_loss.plot()
```

```python
model = Sequential()

model.add(Dense(30,activation='relu'))
model.add(Dropout(0.5)) # half neutron in the above layer of 30 are going to be turned off randomly

model.add(Dense(15,activation='relu'))

model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
early_stop = EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=25)
```

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test),
          callbacks=[early_stop])
```

```python
model_loss = pd.DataFrame(model.history.history)
```

```python
model_loss.plot()
```

```python
predictions = model.predict(X_test)

# if it's binary classification, convert the probabilities to class labels
predicted_classes = (predictions > 0.5).astype("int32")  
```

```python
print(classification_report(y_test,predicted_classes))
```

```python
print(confusion_matrix(y_test,predicted_classes))
```
